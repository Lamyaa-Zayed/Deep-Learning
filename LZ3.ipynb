```
#working with the fashion_mnist data show
import tensorflow as tf
from tensorflow import keras
mnist=tf.keras.datasets.fashion_mnist
(x_images,x_labels),(y_images,y_labels)=mnist.load_data()
x_images.shape
len(x_labels)
y_images.shape
len(y_labels)
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(x_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
class_names=['Top','Trousers','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Boat']
#making normalization on image
x_images=x_images/255.0
y_images=y_images/255.0
plt.figure(figsize=(10,10))
for i in range (36):
  plt.subplot(6,6,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(x_images[i],cmap=plt.cm.binary)
  plt.xlabel(class_names[x_labels[i]])
plt.show()
```

```
#fashion_mnist_with_simple_neural_network(input layed+ one hidden layer+ output layer)
#implementing deep neural network
model=keras.Sequential([keras.layers.Flatten(input_shape=(28,28)), keras.layers.Dense(128,activation='relu'), keras.layers.Dense(10,activation='softmax')])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_images,x_labels, epochs=10)
```

```
#evaluate model on testing parameters
test_loss,test_acc=model.evaluate(x_images,x_labels,verbose=2)
print('\n test accuracy',test_acc)
```

```
#model prediction
import numpy as np
predict=model.predict(y_images)
np.argmax(predict[0])
```
